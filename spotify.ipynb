```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_music = pd.read_csv('music.csv')

display(df_music)
```

# Opgave 1

```python
lower_loudness = df_music['loudness'].quantile(0.25)
upper_loudness = df_music['loudness'].quantile(0.75)
print(lower_loudness, upper_loudness)

speechiness = df_music['speechiness'].quantile(0.25)
print(speechiness)

danciness = df_music['danceability'].quantile(0.75)
print(danciness)

# Find the most popular dance number with less speech content and medium loudness
df_music[(df_music['speechiness'] < speechiness) &
            (df_music['danceability'] > danciness) &
            (df_music['loudness'] > lower_loudness) &
            (df_music['loudness'] < upper_loudness)].sort_values('popularity', ascending=False).head(1)
```

# Opgave 3

```python
import seaborn as sns

def heatmap(df, fig_size=10):
    # Create a heatmap
    corr = df.corr(method="spearman")

    # Mask
    mask = np.triu(np.ones_like(corr, dtype=bool))

    sns.set(font_scale=fig_size / len(corr.columns))

    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    plot, ax = plt.subplots(figsize=(fig_size, fig_size))

    sns.heatmap(corr, vmin=-1, vmax=1, mask=mask,
                cmap=cmap, annot=True, fmt=".2f")

    for axis in [ax.xaxis, ax.yaxis]:
        axis.set(ticks=np.arange(0.5, len(corr.columns)), ticklabels=corr.columns)

    return plot

plot = heatmap(df_music)
```

# Opgave 4

```python
# What genre inbetween between 1999 og 2000 was most popular and what was the tempo in the songs
sorted_df = df_music[(df_music['year'] == 1999) | (df_music['year'] == 2000)]

# Find the most popular genre
genre = sorted_df.groupby('genre')['popularity'].mean().sort_values(ascending=False).head(1)

# Find the tempo of the most popular genre
tempo = sorted_df[sorted_df['genre'] == genre.index[0]]['tempo'].mean()

print("The most popular genres was:", genre.index[0], "and the mean is", tempo)
```

```python
# Find songs, if any, that are explicit but still have considered to have positvie vibes.

# Quantile for valence
valence = df_music['valence'].quantile(0.75)
print(f"The quantile(0.75) for valence: {valence}")

df_music[(df_music['valence'] > valence) &
            (df_music['explicit'] == 1)].sort_values('popularity', ascending=False).head(5)
```

# Opgave 6

```python
# Find the characteristics of sad, uneventful songs in the year 2009
sorted_df = df_music[(df_music['year'] == 2009)]

# Quantile for valence
valence = sorted_df['valence'].quantile(0.25)

# Quantile for energy
energy = sorted_df['energy'].quantile(0.25)

# Quantile for danceability
danceability = sorted_df['danceability'].quantile(0.25)

# Find the songs that are sad, uneventful and have low energy
sorted_df[(sorted_df['valence'] < valence) &
            (sorted_df['energy'] < energy) &
            (sorted_df['danceability'] < danceability)].sort_values('popularity', ascending=False).head(5)
```
